```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

import torchvision
from torchvision import transforms, utils, models

import matplotlib.pyplot as plt
import numpy as np

from functools import partial, reduce
from typing import List
from pytorch_i3d.pytorch_i3d import InceptionI3d
from datasetMVAD import MVADDataset
from transforms import mvad_basic_transform
from extract_feature import Extractor
import os
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda:1')
```

```python
#i3d = InceptionI3d(157, end_module='Mixed_5c')
#i3d.load_state_dict(torch.load('./pytorch_i3d/models/rgb_charades.pt'), strict=False)
i3d = InceptionI3d(400, end_module='Mixed_5c')
i3d.load_state_dict(torch.load('./pytorch_i3d/models/rgb_kinetics.pth'), strict=True)
i3d.to(device)
i3d.eval()
```

```python
MVAD_PATH = '../MVAD'
MVAD_CORPUS_PATH = os.path.join(MVAD_PATH, 'corpus_M-VAD_test.txt')
MVAD_VIDEOS_PATH = os.path.join(MVAD_PATH, 'video')
transform = mvad_basic_transform()
test_dataset = MVADDataset(MVAD_VIDEOS_PATH, MVAD_CORPUS_PATH, train=False, transform=transform)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=True, num_workers=4)
```

```python
MVAD_TEST_SAVE_PATH = os.path.join(MVAD_PATH, 'I3D_rgb', 'test')
```

```python
test_extractor = Extractor(i3d, MVAD_TEST_SAVE_PATH, gpu_device=device)
error_files = test_extractor.extract_feature_from_loader(test_loader)
```
